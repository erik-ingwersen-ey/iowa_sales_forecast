```python
import pandas as pd
from pycaret.regression import *
```

```python
df = pd.read_parquet("../staging/iowa_liquor_daily_sales.parquet")
df
```

```python
s = setup(
    data=df,
    target="Sale (Dollars)",
    categorical_features=["Vendor Number", "City", "Store Number", "Category", "Item Number"],
    polynomial_features=True,
    remove_multicollinearity=True,
    remove_outliers=True,
    transformation=True,
    normalize=True,
    normalize_method="robust",
    # data_split_stratify=True,
)
```

```python
best = compare_models()
```

```python
best
```

```python
evaluate_model(best)
```
